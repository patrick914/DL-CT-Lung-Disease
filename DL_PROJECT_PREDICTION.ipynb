```python
#필요한 Module Import 
import pandas as pd
import tensorflow as tf
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
from glob import glob
```

```python
# Import Google Drive 
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
```

# Model Load

```python
path = '/content/gdrive/MyDrive/2021-2/2학기과목/딥러닝/프로젝트/' #DL_PROJECT_TRAINING.ipynb에서 학습된 모델이 저장된 path 
```

```python
s_model1 = tf.keras.models.load_model(path + 'result/model1.h5') # 학습된 모델 LOAD(1)
s_model2 = tf.keras.models.load_model(path + 'result/model2.h5') # 학습된 모델 LOAD(2)
s_model3 = tf.keras.models.load_model(path + 'result/model3.h5') # 학습된 모델 LOAD(3)
```

# Prediction

```python
# Model1.h5
lc_test_path=os.path.join(path + 'test') # Lung Cancer데이터 path
lc_test_img=glob(os.path.join(lc_test_path,"*.png")) #위의 path에 존재하는 .png파일 LOAD

label_test_1 = pd.read_excel(lc_test_path+'/test_label.xlsx',names = ['name','label'],header=None) # label값이 비워져 있는 파일 load

last_test_datagen=ImageDataGenerator(rescale=1/255.0) #ImageDataGenrator를 통한 전처리
main_generator=last_test_datagen.flow_from_directory(lc_test_path,
                                               batch_size=32,
                                               class_mode='binary',
                                               target_size=(200,200))

result_1 = s_model1.predict(main_generator)<=0.5 #prediction값이 0.5이하라면 True(0) / 0.5 이상이라면 False(1) 기록
label_test_1['label'] = result_1 #비어있던 값에 예측값 저장

label_test_1
```

```python
# Model2.h5
lc_test_path=os.path.join( path + 'test') 
lc_test_img=glob(os.path.join(lc_test_path,"*.png"))

label_test_2 = pd.read_excel(lc_test_path+'/test_label.xlsx',names = ['name','label'],header=None)

last_test_datagen=ImageDataGenerator(rescale=1/255.0)
main_generator=last_test_datagen.flow_from_directory(lc_test_path,
                                               batch_size=32,
                                               class_mode='binary',
                                               target_size=(200,200))

result_2 = s_model2.predict(main_generator)<=0.5
label_test_2['label'] = result_2

label_test_2
```

```python
# Model3.h5

tl_test_path=os.path.join( path + 'test') 
lc_test_img=glob(os.path.join(tl_test_path,"*.png"))

label_test_3 = pd.read_excel(tl_test_path+'/test_label.xlsx',names = ['name','label'],header=None)
last_test_datagen=ImageDataGenerator(rescale=1/255.0)


main_generator = last_test_datagen.flow_from_directory(batch_size=32,
                                                           directory=tl_test_path,
                                                           shuffle=True,
                                                           target_size=(224, 224),
                                                           class_mode='categorical')
result_3 = s_model3.predict(main_generator)<=0.5
label_test_3['label'] = result_3

label_test_3
```

# SAVE RESULT

```python
# 결과를 저장 할 path
result_path = '/content/gdrive/MyDrive/2021-2/2학기과목/딥러닝/프로젝트/result/'
label_test_1.to_csv(result_path+'result_model1.csv')
label_test_2.to_csv(result_path+'result_model2.csv')
label_test_2.to_csv(result_path+'result_model3.csv')
```
